```python
df_tranform = dataset.copy()
```

```python
columns1 = ['gender','partner','dependents','paperlessbilling','churn','phoneservice']
```

```python
for i in columns1:
    if i == 'gender':
        df_tranform[i]=df_tranform[i].map({'Female':0,'Male':1})
    else:
        df_tranform[i]=df_tranform[i].map({'Yes':1,'No':0})
```

```python
columns2 = ['multiplelines','internetservice','onlinesecurity','onlinebackup','deviceprotection','techsupport','streamingtv','streamingmovies','contract','paymentmethod',]
```

```python
df_tranform=pd.get_dummies(df_tranform,columns=columns2)
```

```python
from sklearn.preprocessing import LabelEncoder
lenc = LabelEncoder()
```

```python
df1=df_tranform.copy()
for i in df_tranform.columns:
    df1[i]=lenc.fit_transform(df_tranform[i])
```

```python
df1.head()
```

Logic regression

```python
x=df1.drop('churn',axis=1)
y = df1['churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
churn_model = LogisticRegression()
```

```python
churn_model.fit(X_train,y_train)
```

```python
pred = churn_model.predict(X_test)
pred
```

```python
import sklearn.metrics as sm
accuracy = sm.accuracy_score(y_test,pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix as smat
conf = smat(y_test,pred)
print(conf)
```

```python
from sklearn.metrics import classification_report as cr
print(cr(y_test,pred))
```
